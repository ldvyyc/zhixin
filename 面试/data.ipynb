```python
import pandas as pd
```

```python
df = pd.read_csv('dataset-20230501.csv', header=None)
df
```

```python
# replace Nan with 0
df.fillna(method="ffill", inplace=True)
df
```

```python
# name each column: label, open, high, low, close
df.columns = ['label', 'open', 'high', 'low', 'close']
df
```

```python
df['high_ewma'] = df['high'].ewm(span=10).mean()
df['low_ewma'] = df['low'].ewm(span=10).mean()
df['close_ewma'] = df['close'].ewm(span=10).mean()
df['open_ewma'] = df['open'].ewm(span=10).mean()
df['Volatility'] = df['close'].rolling(window=10).std()
df
```

```python
# drop first 10 rows
df = df[10:]
df
```

```python
df.to_csv('new.csv', index=False)
```

```python
import numpy as np
# only keep rows where there are values = inf
df2 = df[df.isin([np.nan, np.inf, -np.inf]).any(1)]
df2
```

```python
df['close_price'] = (1+df['close']).cumprod()
df
```

```python
df.fillna(1, inplace=True)
df
```

```python
df['label2'] = df['close_price'].shift(-1)
df
```

```python
# 对df进行描述性分析，并画出图表
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('Y_2023.csv')
```

```python
# Plotting TradeVol and Opening in one figure
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(data['Date'], data['TradeVol'], label='TradeVol', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('TradeVol', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Opening'], label='Opening', color='red')
ax2.set_ylabel('Opening', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Trade Number and Opening Number over Time')
plt.show()

# Plotting LPsh, WPsh, and WPsh/LPsh in one figure
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(data['Date'], data['LPsh'], label='LPsh', color='green')
ax1.set_xlabel('Date')
ax1.set_ylabel('LPsh', color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()
ax2.plot(data['Date'], data['WPsh'], label='WPsh', color='blue')
ax2.set_ylabel('WPsh', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Calculating and plotting WPsh/LPsh
wpsh_lpsh_ratio = data['WPsh'] / data['LPsh']
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(data['Date'], wpsh_lpsh_ratio, label='WPsh/LPsh', color='red', linestyle='dashed')
ax3.set_ylabel('WPsh/LPsh', color='red')
ax3.tick_params(axis='y', labelcolor='red')

plt.title('LPsh, WPsh, and WPsh/LPsh over Time')
plt.show()
```
